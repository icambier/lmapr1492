# Téléchargement du fichier CIF symmetrized de MoSe2

```python
with MPRester("h54Ug2jjGzJ0JQCCfHsG") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-7581")
    
    w = CifWriter(structure,symprec=1e-3);
    w.write_file('MoSe2.cif')
structure = Structure.from_file('MoSe2.cif')
```

# Déterminer les 3 premiers pics du diffractogramme correspondant pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å

```python
xrd = XRDCalculator('CuKa')
pattern = xrd.get_pattern(structure = structure, scaled = True, two_theta_range = (0,90))
```

## Premier pic de diffraction

```python
print("Pic de diffraction n°1 : 2 theta = ", pattern.x[0], '°')
print("Intensité ([0,100]): ", pattern.y[0], "%") 
print("Indices hk(i)l : ", pattern.hkls[0][0]["hkl"])
print ("[Note : l'indice i vaut enfait -h - k.]")
print("Multiplicité du pic : ", pattern.hkls[0][0]["multiplicity"])
print("Distance entre les plans : ", pattern.d_hkls[0],"Å")
```

## Deuxième pic de diffraction

```python
print("Pic de diffraction n°2 : 2 theta = ", pattern.x[1], '°')
print("Intensité ([0,100]) : ", pattern.y[1], "%") 
print("Indices hk(i)l : ", pattern.hkls[1][0]['hkl'])
print("Multiplicité du pic : ", pattern.hkls[1][0]['multiplicity'])
print("Distance entre les plans : ", pattern.d_hkls[1],"Å")
```

## Troisième pic de diffraction

```python
print("Pic de diffraction n°3 : 2 theta = ", pattern.x[2], '°')
print("Intensité ([0,100]) : ", pattern.y[2], "%") 
print("Indices hk(i)l : ", pattern.hkls[2][0]['hkl'])
print("Multiplicité du pic : ", pattern.hkls[2][0]['multiplicity'])
print("Distance entre les plans : ", pattern.d_hkls[2],"Å")
```
